```python
data = BPV.Data()
data.read_csv("pixel.dist.csv",False)
data.df.sort_index(by="p",inplace=True,ascending=False)
data.df.set_index(pd.Index([j for j in range(len(data.df))]), inplace=True)
N=50
W=0.09
data.df
```

```python
prbl_exact = BPV.BPV("pulp",data,N,W,time_solver=False)
prbl_exact.solve()
prbl_exact.pprint_solution()
```

```python
prbl_approx = BPV.BPV("heuristic",data,N,W,time_solver=False)
prbl_approx.solve()
prbl_approx.pprint_solution()
#print(BPV.relative_error(prbl_approx,prbl_exact))
heurcost = prbl_approx.heuristic_cost
#np.log2(data.df['p'][heurcost.argmax()])
data.df['p'][heurcost.argmax()]
```

```python
#union
pulp_indexes = data.df[data.df['pulp'] == True].index
pulp_minindex = min(pulp_indexes)
pulp_maxindex = max(pulp_indexes)
heuristic_indexes = data.df[data.df['heuristic'] == True].index
heuristic_minindex = min(heuristic_indexes)
heuristic_maxindex = max(heuristic_indexes)
minidx = min(pulp_minindex,heuristic_minindex)
maxidx = max(pulp_maxindex,heuristic_maxindex)
union = data.df.ix[minidx:maxidx]
union
```
