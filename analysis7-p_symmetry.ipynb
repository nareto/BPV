```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append('/usr/lib/python3.4/site-packages')
import matplotlib
#%matplotlib inline
matplotlib.use('TkAgg')
import BPV
import pattern_manipulation as pm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = BPV.Data()
df.read_csv("p.delviva.csv",False)
#df.df.sort_index(by="p",inplace=True,ascending=False)
df.df.set_index(pd.Index([j for j in range(len(df.df))]), inplace=True)
n = len(df.df)
print(n)
df.df
#df.df[255:]
```

```python
p = df.df['p']
fig = plt.figure()
plt.plot(p)
#plt.show()
```

```python
half_point = int(len(p)/2)
fh =  p[:half_point].copy()
sh = p[half_point:].sort_index(ascending=False).copy()
#sh = p[half_point:].copy()
sh.index = pd.Index(range(half_point))
halfs={'first_half': fh, 'second_half': sh}
phalfs = pd.DataFrame(halfs, columns=['first_half','second_half'])
phalfs
```

```python
fig = plt.figure()
plt.scatter(phalfs['first_half'], phalfs['second_half'])
plt.show()
```

```python
npatterns=30
fig, axes = plt.subplots(npatterns,2,figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})
#fig.subplots_adjust(hspace=0.05, wspace=0)
for i in range(npatterns):
    ax1, ax2 = axes[i]
    ax1.set_title('p = ' + str(df.df['p'][i]))
    ax2.set_title('p = ' + str(df.df['p'][n-1-i]))
    ax1.imshow(df.df['pattern-matrix'][i],cmap=plt.cm.gray,interpolation='nearest')
    ax2.imshow(df.df['pattern-matrix'][n-1-i],cmap=plt.cm.gray,interpolation='nearest')
#for ax, n in zip(axes.flat,range(npatterns)):
#    ax.imshow(,cmap=plt.cm.gray,interpolation='nearest')
plt.show()
```
